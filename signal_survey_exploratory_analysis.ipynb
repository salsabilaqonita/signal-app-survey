```python
numerical = df.select_dtypes(include=['number'])
categorical = df.select_dtypes(exclude=['number'])

# Statistik Deskriptif untuk variabel numerik
print(numerical.describe())
```

```python
# Diagram Batang untuk variabel kategorik
for col in categorical.columns:
    plt.figure(figsize=(6, 3))
    categorical[col].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(f"Distribusi {col}")
    plt.xlabel(col)
    plt.ylabel("Frekuensi")
    plt.xticks(rotation=45)
    plt.show()
```

```python
#Analisis Bivariat
correlation_matrix = numerical.corr()
correlation_matrix
```

```python
# Scatter Plot semua pasangan variabel numerik
sns.pairplot(numerical, diag_kind='kde')
plt.suptitle("Scatter Plot", y=1.02)
plt.show()
```

```python
# Visualisasi distribusi data numerik
df.hist(figsize=(12, 8), bins=20, edgecolor='black')
plt.suptitle("Histogram dari Fitur Numerik")
plt.show()

# Boxplot untuk mendeteksi outlier
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.title("Boxplot dari Fitur Numerik")
plt.show()
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Heatmap Korelasi")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Data Survei Salsa.csv")
print(df.describe())
target_variable = "Mudah"
# Before calculating correlation, filter out non-numeric columns
numerical_df = df.select_dtypes(include=['number'])
# Calculate correlation using the numerical DataFrame
correlation = numerical_df.corr()[target_variable].sort_values(ascending=False)
print("Korelasi dengan variabel:", target_variable)
print(correlation)
plt.figure(figsize=[8, 6])
# Also use the numerical DataFrame for the heatmap
sns.heatmap(numerical_df.corr(), annot=True, cmap="coolwarm", fmt=".2f",
linewidths=0.5)
plt.title("Heatmap Korelasi Variabel dalam Dataset")
plt.show()
```
